```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from pptx import Presentation
from pptx.util import Inches
import seaborn as sns
```

```python
base_dir = "results"
```

```python
def plot_epoch_metrics(logs_path, metric, title, results_path, fname):

    training_logs = pd.read_csv(logs_path)

    fig, ax = plt.subplots()

    # Plot Training loss
    sns.lineplot(x='epoch', y=metric, data=training_logs)

    # Plot Validation loss
    sns.lineplot(x='epoch', y=f"val_{metric}", data=training_logs, color = '#ff7f0e')

    ax.set_title(title)

    # Define custom legend handles
    legend_handles = [
        plt.Line2D([0], [0], color='#1f77b4', lw=2, label='Training'),
        plt.Line2D([0], [0], color='#ff7f0e', lw=2, label='Validation')
    ]

    # Add the legend to the plot
    ax.legend(handles=legend_handles, title="")

    fig.savefig(f"{results_path}/{fname}.png", format='png')
```

```python


subdirs = sorted([d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))])

```

```python
subdirs
```

```python


metrics_path_list = [f"{base_dir}/{subdir}/training_logs.csv" for subdir in subdirs]

```

```python
plot_epoch_metrics(logs_path=metrics_path_list[0],
                   metric="loss", title="EfficientNet B0 (-1)",
                   results_path="paper_figs", fname="b0_minus_1_epoch_loss")

plot_epoch_metrics(logs_path=metrics_path_list[1],
                   metric="loss", title="EfficientNet B0 (-2)",
                   results_path="paper_figs", fname="b0_minus_2_epoch_loss")

plot_epoch_metrics(logs_path=metrics_path_list[2],
                   metric="loss", title="EfficientNet B0 (-3)",
                   results_path="paper_figs", fname="b0_minus_3_epoch_loss")

plot_epoch_metrics(logs_path=metrics_path_list[3],
                   metric="loss", title="EfficientNet B0 (-4)",
                   results_path="paper_figs", fname="b0_minus_4_epoch_loss")

plot_epoch_metrics(logs_path=metrics_path_list[4],
                   metric="loss", title="EfficientNet B0 (-0)",
                   results_path="paper_figs", fname="b0_minus_0_epoch_loss")
```

```python
plot_epoch_metrics(logs_path=metrics_path_list[0],
                   metric="acc", title="EfficientNet B0 (-1)",
                   results_path="paper_figs", fname="b0_minus_1_epoch_acc")

plot_epoch_metrics(logs_path=metrics_path_list[1],
                   metric="acc", title="EfficientNet B0 (-2)",
                   results_path="paper_figs", fname="b0_minus_2_epoch_acc")

plot_epoch_metrics(logs_path=metrics_path_list[2],
                   metric="acc", title="EfficientNet B0 (-3)",
                   results_path="paper_figs", fname="b0_minus_3_epoch_acc")

plot_epoch_metrics(logs_path=metrics_path_list[3],
                   metric="acc", title="EfficientNet B0 (-4)",
                   results_path="paper_figs", fname="b0_minus_4_epoch_acc")

plot_epoch_metrics(logs_path=metrics_path_list[4],
                   metric="acc", title="EfficientNet B0 (-0)",
                   results_path="paper_figs", fname="b0_minus_0_epoch_acc")
```
